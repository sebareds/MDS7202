Para más remate, antes de huir, el practicante les cuenta que este dataframe lamentablemente contiene nuevamente los alumnos de los registros corruptos que ud. y su equipo filtraron en el análisis anterior.
 
 
El rector (evidentemente molesto por la situación) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`). 


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situación (y así evitar buscar duplicados).

> **Hint**: Puede explicar los resultados del merge a través de la cantidad de filas resultantes y los valores que estas contienen.

````python
df_other_grades = pd.read_csv('other_grades.csv')

outer_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="outer")

left_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="left")

right_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="right")

inner_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="inner")

print('Outer Join:')
display(outer_merged_grades)
print('Left Join:')
display(left_merged_grades)
print('Right Join:')
display(right_merged_grades)
print('Inner Join:')
display(inner_merged_grades)
````

> **Justificación:**


El *outer join* deja los datos de ambos datasets, por lo que los nombres "erróneos" (que no están en el dataset original limpiado) se agregan igual, incluyendo solamente las notas del segundo dataset (puntaje de historia y ciencias), y aumentando así el número de filas a 1000.

El *inner join* deja a las personas que estén en ambos datasets, por lo que en este caso, como el nuevo dataset encontrado contiene a todas las personas del dataset original (limpiado) y otras erróneas, se borran las erróneas y se agregan las notas de los 5 ramos para las personas del dataset original (manteniéndose las 875).

El *left join* (aplicando el dataset original a la izquierda) dejará a todas las personas del dataset original junto a las 3 notas originales, mientras que agregará las notas de historia y ciencias para las que, además de estar en el dataset original, están en el nuevo dataset encontrado (a las que no, les rellena estas notas con NaN). Como ya dijimos, todas las personas del dataset original están en el nuevo, por lo que el *left join* funciona como el *inner join*.

El *right join* (aplicando el dataset original a la izquierda) dejará a todas las personas del dataset nuevo, junto a sus notas de historia y ciencias, y agregará las notas de matemáticas, lectura y escritura para las personas del dataset nuevo que también están en el original (a las que no, les rellena estas notas con NaN). Como ya dijimos, todas las personas del dataset original están en el nuevo, por lo todas ellas quedarán con sus 5 notas, y el resto quedará con las de historia y ciencias, es decir, funciona igual al *outer join*.


Dado lo anterior, para este caso serviría tanto el *left join* (si se ocupa el dataset original a la izquierda, y si se ocupa a la derecha, serviría el *right join*) como el *inner join* para generar el nuevo DataFrame. Eso sí, para un caso un poco más general, dependería de si se quieren mantener solamente los que tengan notas de los 5 ramos (en este caso habría que ocupar *inner join*) o si se quiere mantener de todas maneras a las personas que estaban en el dataset original limpiado a pesar de que en el nuevo dataset encontrado (el de las notas de historia y ciencias) no estén sus notas (en este caso habría que ocupar *left join*, tomando el dataset original a la izquierda, o *right join* con el dataset original a la derecha, pero cabe notar que en este problema esto no pasa, pues el nuevo dataset encontrado contiene a todos las personas del dataset original limpiado).
