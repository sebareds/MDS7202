````python
df_other_grades = pd.read_csv('other_grades.csv')

outer_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="outer")

left_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="left")

right_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="right")

inner_merged_grades = pd.merge(left=df_grades,
                           right=df_other_grades,
                           on="names",
                           how="inner")

print('Outer Join:')
display(outer_merged_grades)
print('Left Join:')
display(left_merged_grades)
print('Right Join:')
display(right_merged_grades)
print('Inner Join:')
display(inner_merged_grades)
````

> **Justificaci칩n:**


El *outer join* deja los datos de ambos datasets, por lo que los nombres "err칩neos" (que no est치n en el dataset original limpiado) se agregan igual, incluyendo solamente las notas del segundo dataset (puntaje de historia y ciencias), y aumentando as칤 el n칰mero de filas a 1000.

El *inner join* deja a las personas que est칠n en ambos datasets, por lo que en este caso, como el nuevo dataset encontrado contiene a todas las personas del dataset original (limpiado) y otras err칩neas, se borran las err칩neas y se agregan las notas de los 5 ramos para las personas del dataset original (manteni칠ndose las 875).

El *left join* (aplicando el dataset original a la izquierda) dejar치 a todas las personas del dataset original junto a las 3 notas originales, mientras que agregar치 las notas de historia y ciencias para las que, adem치s de estar en el dataset original, est치n en el nuevo dataset encontrado (a las que no, les rellena estas notas con NaN). Como ya dijimos, todas las personas del dataset original est치n en el nuevo, por lo que el *left join* funciona como el *inner join*.

El *right join* (aplicando el dataset original a la izquierda) dejar치 a todas las personas del dataset nuevo, junto a sus notas de historia y ciencias, y agregar치 las notas de matem치ticas, lectura y escritura para las personas del dataset nuevo que tambi칠n est치n en el original (a las que no, les rellena estas notas con NaN). Como ya dijimos, todas las personas del dataset original est치n en el nuevo, por lo todas ellas quedar치n con sus 5 notas, y el resto quedar치 con las de historia y ciencias, es decir, funciona igual al *outer join*.


Dado lo anterior, para este caso servir칤a tanto el *left join* (si se ocupa el dataset original a la izquierda, y si se ocupa a la derecha, servir칤a el *right join*) como el *inner join* para generar el nuevo DataFrame. Eso s칤, para un caso un poco m치s general, depender칤a de si se quieren mantener solamente los que tengan notas de los 5 ramos (en este caso habr칤a que ocupar *inner join*) o si se quiere mantener de todas maneras a las personas que estaban en el dataset original limpiado a pesar de que en el nuevo dataset encontrado (el de las notas de historia y ciencias) no est칠n sus notas (en este caso habr칤a que ocupar *left join*, tomando el dataset original a la izquierda, o *right join* con el dataset original a la derecha, pero cabe notar que en este problema esto no pasa, pues el nuevo dataset encontrado contiene a todos las personas del dataset original limpiado).
